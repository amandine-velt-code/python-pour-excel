## Librairie XlsxWriter

```python
import pandas as pd

# Petit tableau de démonstration
df = pd.DataFrame({
    "Mois": ["Janvier", "Février"],
    "Ventes": [1200, 1500]
})

# Création d'un fichier Excel simple
with pd.ExcelWriter("../data/rapport_automatique.xlsx", engine="xlsxwriter") as writer:
    df.to_excel(writer, index=False)

    workbook = writer.book
    worksheet = writer.sheets["Sheet1"]

    # Un format très simple : en-têtes avec fond jaune clair
    format_header = workbook.add_format({
        "bold": True,
        "bg_color": "#FFF2CC"   # jaune pastel
    })

    # On applique le style aux en-têtes
    worksheet.write("A1", "Mois", format_header)
    worksheet.write("B1", "Ventes", format_header)

    # On élargit légèrement les colonnes (pour que ce soit joli)
    worksheet.set_column("A:B", 12)
```

## Librairie xlwings

```python
import xlwings as xw

# dans le fichier d'origine, la première ligne (en-tête) est : Rapport Mensuel des Ventes — Janvier 2025
# on modifie automatiquement en : Rapport Mensuel des Ventes — Février 2025
wb = xw.Book("../data/rapport.xlsx")
sheet = wb.sheets["Rapport_vente"]
sheet["A1"].value = "Rapport Mensuel des Ventes — Février 2025"
```
